# L1 Objectstrings

Qualtran defines bloq classes, but we need to provide compile-time classical parameters to instantiate those into bloq objects. 

In principle, arbitrary Python values can be used as classical parameters, as long as they are immutable and hashable. For a human readable intermediate representation, `qualtran.l1` implements a limited serialization syntax called **objectstrings** that roughly mimics standard Python object instantiation.

## Example Usage

### Loading

```python
from qualtran.l1 import load_objectstring
load_objectstring('QInt(8)')
```

### Dumping

We'll create the same 8-bit quantum integer programatically.

```python
import qualtran as qlt
import qualtran.dtype as qdt

qint8 = qdt.QInt(8)
```

It can be dumped to an objectstring.

```python
from qualtran.l1 import dump_objectstring

dump_objectstring(qint8)
```

### Bloqs

```python
load_objectstring('qualtran.bloqs.arithmetic.AddK(k=5, dtype=QUInt(4))')
```

## Motivation: Compile-Time Classical Parameters

A bloq *object* is a particular bloq *class* with a collection of compile-time classical parameters. For example, if we want to add a constant to a quantum register we instantiate an AddK bloq:

```python
from qualtran.bloqs.arithmetic import AddK
add_five = AddK(k=5, dtype=qdt.QInt(8))
add_five
```

AddK is a bloq class (i.e. a Python class with name "AddK") and we provide the compile-time classical arguments `k=5` and `dtype=QInt` to configure the family of quantum circuits to specifically the ones that represent adding the classical constant "5" to a quantum register holding an 8-qubit quantum integer. We assign the bloq object to the variable named `add_five`. 

## Advanced concepts

### AST Nodes

```python
from qualtran.l1 import to_cobject_node
```

Loading and dumping both go via an abstract syntax tree (AST) representation. This can be used directly. For example, any Python object will be converted to a `LiteralNode`, `TupleNode`, or `ObjectNode`.

```python
print(to_cobject_node(5))
print(to_cobject_node((5, 'hello')))
print(to_cobject_node(qdt.QInt(8)))
```

### Canonicalization
The AST nodes have a method for dumping a canonical string representation. The objectstring format has a deliberately small set of container classes to ensure canonical string representations.

```python
to_cobject_node(qdt.QInt(bitsize=8)).canonical_str()
```

```python
load_objectstring('MyStruct(5)')
```

### Singletons

A singleton object or identifier is indistinguishable from an object with no arguments. The canonical form of an object constructed with no arguments omits the traditional `()` constructor call indication found in Python.

```python
dump_objectstring(qdt.QBit())
```

```python
load_objectstring('QBit()') == load_objectstring('QBit')
```

```python
dump_objectstring(True)
```
