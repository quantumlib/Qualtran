# BloqBuilder Gate Helper Methods

For convenience, `qualtran.BloqBuilder` has methods for adding ("calling") common basic gates to your program. Each method maps 1-to-1 to a bloq class, particularly the `.qcall(...)` class method. 

```python
# Top-level qualtran imports
import qualtran as qlt
import qualtran.dtype as qdt
```

### Single qubit gates

```python
bb = qlt.BloqBuilder()
q = bb.in_register('q', qdt.QBit())

q = bb.X(q)
q = bb.Y(q)
q = bb.Z(q)
q = bb.H(q)
q = bb.T(q)
q = bb.T(q, is_adjoint=True)
q = bb.S(q)
q = bb.S(q, is_adjoint=True)
q = bb.identity(q)

qlt.show_bloq(bb.finalize(q=q))
```

### Two qubit gates

```python
bb = qlt.BloqBuilder()
q1 = bb.in_register('q1', qdt.QBit())
q2 = bb.in_register('q2', qdt.QBit())

q1, q2 = bb.CNOT(q1, q2)
q1, q2 = bb.CZ(q1, q2)
q1, q2 = bb.CY(q1, q2)
q1, q2 = bb.CH(q1, q2)
q1, q2 = bb.swap(q1, q2)
[q1, q2], anc = bb.And([q1, q2])
[q1, q2] = bb.UnAnd([q1, q2], anc)

qlt.show_bloq(bb.finalize(q1=q1, q2=q2))
```

### Three qubit gates

```python
bb = qlt.BloqBuilder()
q1 = bb.in_register('q1', qdt.QBit())
q2 = bb.in_register('q2', qdt.QBit())
q3 = bb.in_register('q3', qdt.QBit())

q1, q2, q3 = bb.Toffoli(q1, q2, q3)
q1, q2, q3 = bb.cswap(q1, q2, q3)

qlt.show_bloq(bb.finalize(q1=q1, q2=q2, q3=q3))
```

### Rotations

```python
bb = qlt.BloqBuilder()
q1 = bb.in_register('q1', qdt.QBit())
q2 = bb.in_register('q2', qdt.QBit())

q1 = bb.rx(q1, 0.1)
q1 = bb.ry(q1, 0.2)
q1 = bb.rz(q1, 0.3)
q1, q2 = bb.cry(q1, q2, 0.4)
q1, q2 = bb.crz(q1, q2, 0.5)
q1 = bb.su2_rotation(q1, theta=0.1, phi=0.2, lambd=0.3)
q1 = bb.z_pow(q1, exponent=0.6)
q1, q2 = bb.cz_pow(q1, q2, exponent=0.7)
q1 = bb.x_pow(q1, exponent=0.8)
q1 = bb.y_pow(q1, exponent=0.9)

qlt.show_bloq(bb.finalize(q1=q1, q2=q2))
```

### Measurement and discard

```python
bb = qlt.BloqBuilder()
q1 = bb.in_register('q1', qdt.QBit())
q2 = bb.in_register('q2', qdt.QBit())

c1 = bb.measure_x(q1)
c2 = bb.measure_z(q2)

bb.discard(c1)
bb.discard(c2)

q3 = bb.alloc_qbit()
bb.discard_q(q3)

qlt.show_bloq(bb.finalize())
```

### Allocations

```python
bb = qlt.BloqBuilder()

q = bb.alloc_qbit(0)
bb.free_qbit(q, 0)

x = bb.alloc_qint(55, bitsize=8)
bb.free(x)

y = bb.allocate(3)
bb.free(y)

qlt.show_bloq(bb.finalize())
```

### Bookkeeping

```python
bb = qlt.BloqBuilder()
x = bb.in_register('x', qdt.QAny(3))

xs = bb.split(x)
x = bb.join(xs)

qlt.show_bloq(bb.finalize(x=x))
```
