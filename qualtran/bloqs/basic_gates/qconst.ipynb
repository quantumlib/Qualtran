# Quantum Constants

```python
from qualtran import Bloq, CompositeBloq, BloqBuilder, Signature, Register
from qualtran import QBit, QInt, QUInt, QAny
from qualtran.drawing import show_bloq, show_call_graph, show_counts_sigma
from typing import *
import numpy as np
import sympy
import cirq
```

## `QUIntState`
The state |val> for an unsigned integer val of type QUInt(bitsize)

#### Parameters
 - `val`: the classical value
 - `bitsize`: The bitsize of the register 

#### Registers
 - `val`: The register of size `bitsize` which initializes the value `val`.


```python
from qualtran.bloqs.basic_gates import QUIntState
```

### Example Instances

```python
quint_state = QUIntState(5, bitsize=8)
```

#### Graphical Signature

```python
from qualtran.drawing import show_bloqs
show_bloqs([quint_state],
           ['`quint_state`'])
```

### Call Graph

```python
from qualtran.resource_counting.generalizers import ignore_split_join
quint_state_g, quint_state_sigma = quint_state.call_graph(max_depth=1, generalizer=ignore_split_join)
show_call_graph(quint_state_g)
show_counts_sigma(quint_state_sigma)
```
